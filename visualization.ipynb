```python
import pandas as pd
from matplotlib.pyplot import *
%matplotlib inline
```

```python
x=pd.read_csv("C:/Users/Akshat Aman/Desktop/DE.csv" , header = None)
type(x)
x.head()
```

```python
fig, ax = subplots()
df = x.plot( kind = 'line',ax = ax, sharey = True, legend = True, )
ax.legend(["Validation","Training"])
df.set_xlabel("Epochs")
df.set_ylabel("Loss (in percentage)")
fig.savefig("ModelLoss")
```
