```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
import keras
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.models import Sequential
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
irregular = cv2.imread("./Category1/Irregular/download (2).jpg")
```

```python
plt.imshow(irregular)
```

```python
irregular.shape
```

```python
num_classes=3
```

```python
model1 = Sequential()
model1.add(Conv2D(3,5,5, activation='tanh', input_shape=irregular.shape))
```

```python
def visualize_irr(model, irregular):
    fig = plt.gcf()
    irr_batch = np.expand_dims(irregular,axis=0)
    conv_irr = model.predict(irr_batch)
    conv_irr = np.squeeze(conv_irr, axis=0)
    print (conv_irr.shape)
    plt.imshow(conv_irr)
    fig.savefig('3_filter_1st_layer_visuals.png')
```

```python
visualize_irr(model1, irregular)
```

```python
model2 = Sequential()
model2.add(Conv2D(1,5,5, activation='tanh', input_shape=irregular.shape))
model2.add(MaxPooling2D(pool_size=(5,5)))
```

```python
def nice_galaxy_printer(model, irregular):
    fig = plt.gcf()
    irr_batch = np.expand_dims(irregular,axis=0)
    conv_irr2 = model.predict(irr_batch)

    conv_irr2 = np.squeeze(conv_irr2, axis=0)
    print (conv_irr2.shape)
    conv_irr2 = conv_irr2.reshape(conv_irr2.shape[:2])

    print (conv_irr2.shape)
    plt.imshow(conv_irr2)
    fig.savefig('1_filter_1st_layer_visuals_with_pooling.png')
```

```python
nice_galaxy_printer(model2, irregular)
```

```python
model3 = Sequential()
model3.add(Conv2D(3,5,5, activation='tanh', input_shape=irregular.shape))
model3.add(MaxPooling2D(pool_size=(5,5)))
model3.add(Conv2D(3,5,5, activation='tanh', input_shape=irregular.shape))
model3.add(MaxPooling2D(pool_size=(5,5)))

visualize_irr(model3, irregular)
```
